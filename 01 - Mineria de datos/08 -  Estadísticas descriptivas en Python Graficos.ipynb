
# Histograma

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np

# Cambiamos el directorio local a donde tenemos nuestras bases de datos
os.chdir('BasesDatos/')
```

```python
# Cargar la base de datos con la que se va a trabajar
df_ventas = pd.read_excel('ventas.xlsx')
```

```python
df_ventas.head()
```

```python
# listar los estilos de graficos disponibles
print(plt.style.available)
```

```python
# Seleccionar estilo de grafico ggplot
plt.style.use('ggplot')

# Contruir un histograma
plt.hist(df_ventas.loc[df_ventas["PrecioUnidad"]>300, "PrecioUnidad"], color='r', bins = 10)

# poner nombre al eje Y
plt.ylabel('Total de casos')

# Poner titulo a la grafica
plt.title('Histograma del total precio de la unidad')

# Agregar comentarios al grafico
plt.text(10000, 100, r'prueba de texto $\mu$')

" Activar la regilla "
plt.grid(True)

# Seleccionar la scala del eje Y
plt.yscale('linear')


# Mostrar el grafico
plt.show()
```

```python
# Contruir el grafico anterior en seaborn
sns.distplot( df_ventas[df_ventas["PrecioUnidad"]>300]["PrecioUnidad"].astype(float),  kde=False)
plt.show()
```

# Grfico de lineas

```python
t = np.arange(0.1, 20, 0.1)

y1 = np.sin(t)/t
y2 = np.sin(t)*np.exp(-t)
p1, p2 = plt.plot(t, y1, t, y2)

# Texto en la grfica en coordenadas (x,y)
texto1 = plt.text(2, 0.6, r'$\frac{\sin(x)}{x}$', fontsize=20)
texto2 = plt.text(13, 0.2, r'$\sin(x) \cdot e^{-x}$', fontsize=16)

plt.title('Representacion de dos funciones')
plt.xlabel('Tiempo / s')
plt.ylabel('Amplitud / cm')

plt.show()
```
